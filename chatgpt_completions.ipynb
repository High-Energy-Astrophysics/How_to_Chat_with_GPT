```python
'''
Generate Python code that solves this task:
Generate 2 distinct datagroups each with 100 datapoints using a 2D normal distribution and plot them. 
Stack the data into single array and use K-means clustering algorithm to divide them. 
Estimate how many datapoints were clustered incorrectly and plot them.
'''

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate two groups of 2D data using normal distribution
group1 = np.random.multivariate_normal(mean=[1, 1], cov=[[1, 0], [0, 1]], size=100)
group2 = np.random.multivariate_normal(mean=[-1, -1], cov=[[1, 0], [0, 1]], size=100)

# Plot the generated data
plt.scatter(group1[:, 0], group1[:, 1], color='blue')
plt.scatter(group2[:, 0], group2[:, 1], color='red')
plt.title('Original Data')
plt.show()

# Stack the data into a single array
data = np.vstack((group1, group2))

# Apply K-means clustering algorithm with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=0).fit(data)

# Predict the labels for the data points
labels = kmeans.predict(data)

# Count the number of incorrectly clustered data points
incorrectly_clustered = np.sum(labels == np.array([0]*100 + [1]*100))

# Get the coordinates of incorrectly clustered data points
incorrect_points = data[labels == np.array([0]*100 + [1]*100)]

# Plot the clustered data points with different colors
plt.scatter(data[labels==0, 0], data[labels==0, 1], color='blue')
plt.scatter(data[labels==1, 0], data[labels==1, 1], color='red')
plt.title(f'Clustered Data ({incorrectly_clustered} points incorrectly clustered)')

# Plot incorrectly clustered data points with black crosses
plt.scatter(incorrect_points[:, 0], incorrect_points[:, 1], marker='x', color='black')

plt.show()
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Generate 100 random numbers
random_numbers = [random.randint(0, 100) for i in range(100)]

# Plot histogram
plt.hist(random_numbers, bins=10)
plt.title("Histogram of 100 Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Calculate skewness
skewness = skew(random_numbers)
print("The skewness of the distribution is:", skewness)
```
